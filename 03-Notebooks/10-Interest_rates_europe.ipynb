```python
import requests
import pandas as pd
import io
from pandas.tseries.offsets import MonthEnd
```

```python
# Building blocks for the URL
entrypoint = 'https://sdw-wsrest.ecb.europa.eu/service/' # Using protocol 'https'
resource = 'data'           # The resource for data queries is always'data'
flowRef ='MIR'              # Dataflow describing the data that needs to be returned, exchange rates in this case
key = 'M.U2.B.A20.A.R.A.2240.EUR.O'    # Defining the dimension values, explained below

# Define the parameters
'''parameters = {
    'startPeriod': '2000-01-01',  # Start date of the time series
    'endPeriod': '2018-10-01'     # End of the time series
}'''

request_url = entrypoint + resource + '/'+ flowRef + '/' + key

response = requests.get(request_url, headers={'Accept': 'text/csv'})

interest_rates = pd.read_csv(io.StringIO(response.text))

interest_rates = interest_rates[['TIME_PERIOD', 'OBS_VALUE']]

interest_rates.rename(columns={
    'TIME_PERIOD': 'Date',
    'OBS_VALUE': 'Value'
}, inplace=True)

interest_rates.set_index('Date', inplace=True)

interest_rates.index = pd.to_datetime(interest_rates.index, format='%Y-%m')

interest_rates.index = pd.to_datetime(interest_rates.index, format='%Y') + MonthEnd(1)

interest_rates
```
