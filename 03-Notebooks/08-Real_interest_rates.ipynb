### Tasa de interés real = Tasa de interés nominal - Inflación

### Datos mensuales

```python
inflation_monthly = pd.read_pickle(r'..\02-Data\05-Inflation\01_Inflation_percentage_1913-01-01_2021-06-01_monthly_fred.pkl')
interest_rate_monthly = pd.read_pickle(r'..\02-Data\07-Interest_rates\02-InterestRates_percentage_1954-07-01_2021-07-22_monthly_fred.pkl')
```

```python
inflation_monthly
```

```python
interest_rate_monthly
```

```python
mask = (inflation_monthly.index >= datetime.strptime('1954-07-01', '%Y-%m-%d'))
inflation_monthly= inflation_monthly.loc[mask]
inflation_monthly
```

```python
real_interest_rate_monthly = interest_rate_monthly.set_index(inflation_monthly.index).subtract(inflation_monthly, fill_value=0)
```

```python
real_interest_rate_monthly
```

```python
real_interest_rate_monthly.to_pickle(r'..\02-Data\08-Real_interest_rates\01-RealInterestRates_percentage_1954-07-01_2021-06-01_monthly_fred.pkl')
```

```python
real_interest_rate_monthly.to_parquet(r'..\02-Data\08-Real_interest_rates\01-RealInterestRates_percentage_1954-07-01_2021-06-01_monthly_fred.parquet')
```
